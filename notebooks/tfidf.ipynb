```python
remove_digits = str.maketrans('', '', digits)
```

```python
df.preprocessed_text = df.preprocessed_text.apply(lambda x: x.translate(remove_digits))
```

```python
X = df.preprocessed_text.drop_duplicates().dropna().tolist()
```

```python
len(X)
```

```python
countvectorizer = CountVectorizer(ngram_range=(1,2))
```

```python
X1 = countvectorizer.fit_transform(X) 
features = (countvectorizer.get_feature_names_out())
```

```python
features.tolist()
```

# applying tfidf

```python
vectorizer = TfidfVectorizer(ngram_range = (1,2))
X2 = vectorizer.fit_transform(X)
scores = (X2.toarray())
print("\n\nScores : \n", scores)
```

```python
sums = X2.sum(axis = 0)
data1 = []
for col, term in enumerate(features):
    data1.append( (term, sums[0,col] ))
ranking = pd.DataFrame(data1, columns = ['term','rank'])
words = (ranking.sort_values('rank', ascending = False))
print ("\n\nWords head : \n", words.head(100))
```
