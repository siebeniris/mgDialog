```python
import pandas as pd
import numpy as np
import json
from glob import glob
from collections import defaultdict
import json
import tqdm
```

```python
topic_eu_dict=defaultdict(dict)

for file in tqdm.tqdm(glob(f"../output/tp/eu/**/**_etm.csv", recursive=True)):
    print(file)
    df = pd.read_csv(file)
    
    lang = file.split("/")[-1].replace("_etm.csv","")
    query=file.split("/")[3]
    print(lang, query)
    
    df["year"]= df["date"].str[:4]
    counter = df.groupby(["topic","year"]).agg("count").to_dict()["resourceId"]
    
    for idx, count in counter.items():
        topic_nr, year = idx
        topic_eu_dict[f"{lang}_{topic_nr}"][year]= count
    del df
    
    
```

```python
df_eu_topic = pd.DataFrame.from_dict(topic_eu_dict)
```

```python
topic2words = defaultdict(dict)

for query in ["eu", "un"]:
    topic2words[query] = defaultdict(dict)
    for file in glob(f"../output/tp/{query}/**/topic_words.json", recursive=True):
        
        print(file)
        lang = file.split("/")[4]
        topic2words[query][lang] = defaultdict(dict)
        print(lang)
        
        with open(file) as f:
            topicwords = json.load(f)
            
        df_tfidf = pd.read_csv(f"../output/tfidf/{query}/{lang}.csv")
        
        lang_tfidf = dict(zip(df_tfidf["term"], df_tfidf["rank"]))
        
        for idx, wordlist in topicwords.items():
            word2rank= {x:lang_tfidf.get(x,np.NaN) for x in wordlist}
            sorted_word2rank = dict(sorted(word2rank.items(), reverse=True, key=lambda x:x[1]))
            
            topic2words[query][lang][idx]=sorted_word2rank
        
```
