```python
folder = "data/raw/un/de"
stats_dict=defaultdict(list)
query_names = []
domains =[]
pubTypes = []

for file in os.listdir(folder):
    if file.endswith(".csv"):
        filepath = os.path.join(folder, file)
        df = pd.read_csv(filepath, index_col=0, low_memory=False)
        if "countryCode" in df.columns:
            stats_dict["countryCode"]+=df["countryCode"].dropna().tolist()
        if "queryName" in df.columns:
            query_names += df["queryName"].dropna().tolist()
        if "domain" in df.columns:
            domains +=df["domain"].dropna().tolist()
        if "pageTypeName" in df.columns:
            pubTypes += df["pageTypeName"].dropna().tolist()
```

```python
langs = ["cs", "da", "de", "el", "en", "es", "fr", "hu", "it", "nl", "pl", "sk", "sv"]
```

```python
df_list =[]
for lang in langs:
    filepath = f"output/non-nas/eu_{lang}.csv"
    df = pd.read_csv(filepath, sep="~", header=None, index_col=0)
    df = df.groupby(df.index).agg(sum)
    df.columns=[lang]
    df_list.append(df)
```

```python
df_list[0]
```

```python
df_un = pd.concat(df_list, axis=1)
```

```python
df_un
```

```python
df_un.to_csv("output/non-nas/eu_nanan.csv")
```
